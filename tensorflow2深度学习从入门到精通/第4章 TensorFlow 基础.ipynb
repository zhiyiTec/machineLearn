```python
# 4.2 æ•°å€¼ç²¾åº¦
import tensorflow as tf
import numpy as np
a=tf.constant(123456789, dtype=tf.int16)
a1=tf.constant(123456789, dtype=tf.int32)
print(a)
print(a1)
a2=tf.constant(np.pi, dtype=tf.float32)
a3=tf.constant(np.pi, dtype=tf.float64)
print(a2)
print(a3)
## 4.2.1 è¯»å–ç²¾åº¦
print('before:',a.dtype)
if a.dtype != tf.float32:
    a = tf.cast(a,tf.float32) # è½¬æ¢ç²¾åº¦
print('after :',a.dtype)
## 4.2.2 ç±»å‹è½¬æ¢
a = tf.constant(np.pi, dtype=tf.float16)
a=tf.cast(a, tf.double)
print("aè½¬double:{}".format(a.dtype))
### è¿›è¡Œç±»å‹è½¬æ¢æ—¶ï¼Œéœ€è¦ä¿è¯è½¬æ¢æ“ä½œçš„åˆæ³•æ€§ï¼Œä¾‹å¦‚å°†é«˜ç²¾åº¦çš„å¼ é‡è½¬æ¢ä¸ºä½ç²¾åº¦çš„å¼ é‡æ—¶ï¼Œå¯èƒ½å‘ç”Ÿæ•°æ®æº¢å‡ºéšæ‚£ï¼š
a=tf.constant(123456789,dtype=tf.int32)
a=tf.cast(a,tf.int16)
print("aæœ‰32è½¬16ä½:",a)
### å¸ƒå°”å‹ä¸æ•´å½¢ä¹‹é—´ç›¸äº’è½¬æ¢ä¹Ÿæ˜¯åˆæ³•çš„ï¼Œæ˜¯æ¯”è¾ƒå¸¸è§çš„æ“ä½œï¼š
a = tf.constant([True, False])
a=tf.cast(a,tf.int32)
print("å¸ƒå°”å‹ä¸æ•´å½¢ä¹‹é—´ç›¸äº’è½¬æ¢ï¼š",a)
### ä¸€èˆ¬é»˜è®¤ 0 è¡¨ç¤º Falseï¼Œ1 è¡¨ç¤º Trueï¼Œåœ¨ TensorFlow ä¸­ï¼Œå°†é 0 æ•°å­—éƒ½è§†ä¸º True
a = tf.constant([-1, 0, 1, 2])
a=tf.cast(a, tf.bool)
print("æ•´å‹è½¬boolç±»å‹ï¼š",a)
```

```python
# 4.4 åˆ›å»ºå¼ é‡
### Numpy Array æ•°ç»„å’Œ Python List æ˜¯ Python ç¨‹åºä¸­é—´éå¸¸é‡è¦çš„æ•°æ®è½½ä½“å®¹å™¨
### å¾ˆå¤šæ•°æ®éƒ½æ˜¯é€šè¿‡ Python è¯­è¨€å°†æ•°æ®åŠ è½½è‡³ Array æˆ–è€… List å®¹å™¨ï¼Œå†è½¬æ¢åˆ° Tensor ç±»å‹
### é€šè¿‡TensorFlow è¿ç®—å¤„ç†åå¯¼å‡ºåˆ° Array æˆ–è€… List å®¹å™¨ï¼Œæ–¹ä¾¿å…¶ä»–æ¨¡å—è°ƒç”¨ã€‚
## 4.4.1 ä» ä» Numpy, List å¯¹è±¡ åˆ›å»º
### é€šè¿‡ tf.convert_to_tensor å¯ä»¥åˆ›å»ºæ–° Tensorï¼Œå¹¶å°†ä¿å­˜åœ¨ Python List å¯¹è±¡æˆ–è€… NumpyArray å¯¹è±¡ä¸­çš„æ•°æ®å¯¼å…¥åˆ°æ–° Tensor ä¸­
a=tf.convert_to_tensor([1,2.])
### éœ€è¦æ³¨æ„çš„æ˜¯ï¼ŒNumpy ä¸­æµ®ç‚¹æ•°æ•°ç»„é»˜è®¤ä½¿ç”¨ 64-Bit ç²¾åº¦ä¿å­˜æ•°æ®ï¼Œ
### è½¬æ¢åˆ° Tensor ç±»å‹æ—¶ç²¾åº¦ä¸º tf.float64ï¼Œå¯ä»¥åœ¨éœ€è¦çš„æ—¶å€™è½¬æ¢ä¸º tf.float32 ç±»å‹ã€‚
aa=tf.convert_to_tensor(np.array([[1,2.],[3,4]]))
print("a:",a)
print("aa:",aa)
## 4.4.2 åˆ›å»ºå…¨ å…¨ 0 ï¼Œå…¨ 1 å¼ é‡
b=tf.zeros([])
b1=tf.ones([])
b2=tf.ones([1])
b3=tf.zeros([1])
print("b:",b)
print("b1",b1)
print("b2",b2)
print("b3",b3)
# åˆ›å»ºå…¨ 0 çš„çŸ©é˜µï¼š
b4=tf.zeros([2,3])
print("b4",b4)
# åˆ›å»ºå…¨1çŸ©é˜µ
b5=tf.ones([6,6])
print("b5",b5)
### è¿‡ tf.zeros_like, tf.ones_like å¯ä»¥æ–¹ä¾¿åœ°æ–°å»ºä¸æŸä¸ªå¼ é‡ shape ä¸€è‡´ï¼Œå†…å®¹å…¨ 0 æˆ–å…¨ 1çš„å¼ é‡ã€‚
### ä¾‹å¦‚ï¼Œåˆ›å»ºä¸å¼ é‡ b6 å½¢çŠ¶ä¸€æ ·çš„å…¨ 0 å¼ é‡ï¼š
b6=tf.ones([2,3])
b7=tf.zeros_like(b6)
print("b7",b7)
    ## 4.4.3 åˆ›å»º åˆ›å»º è‡ªå®šä¹‰æ•°å€¼
### é™¤äº†åˆå§‹åŒ–ä¸ºå…¨ 0ï¼Œæˆ–å…¨ 1 çš„å¼ é‡ä¹‹å¤–ï¼Œæœ‰æ—¶ä¹Ÿéœ€è¦å…¨éƒ¨åˆå§‹åŒ–ä¸ºæŸä¸ªè‡ªå®šä¹‰æ•°å€¼çš„å¼ é‡ï¼Œæ¯”å¦‚å°†å¼ é‡çš„æ•°å€¼å…¨éƒ¨åˆå§‹åŒ–ä¸º-1 ç­‰ã€‚
### é€šè¿‡ tf.fill(shape, value)å¯ä»¥åˆ›å»ºå…¨ä¸ºè‡ªå®šä¹‰æ•°å€¼ value çš„å¼ é‡ã€‚ä¾‹å¦‚ï¼Œåˆ›å»ºå…ƒç´ ä¸º-1çš„æ ‡é‡ï¼š
c=tf.fill([3,4],-1)
print("C:",c)
## 4.4.4 åˆ›å»ºå·²çŸ¥åˆ†å¸ƒçš„å¼ é‡
### æ­£æ€åˆ†å¸ƒ(Normal Distributionï¼Œæˆ– Gaussian Distribution)
### å‡åŒ€åˆ†å¸ƒ(UniformDistribution)æ˜¯æœ€å¸¸è§çš„åˆ†å¸ƒä¹‹ä¸€ï¼Œåˆ›å»ºé‡‡æ ·è‡ªè¿™ 2 ç§åˆ†å¸ƒçš„å¼ é‡éå¸¸æœ‰ç”¨
### æ¯”å¦‚åœ¨å·ç§¯ç¥ç»ç½‘ç»œä¸­ï¼Œå·ç§¯æ ¸å¼ é‡ W åˆå§‹åŒ–ä¸ºæ­£æ€åˆ†å¸ƒæœ‰åˆ©äºç½‘ç»œçš„è®­ç»ƒï¼›
### åœ¨å¯¹æŠ—ç”Ÿæˆç½‘ç»œä¸­ï¼Œéšè—å˜é‡ z ä¸€èˆ¬é‡‡æ ·è‡ªå‡åŒ€åˆ†å¸ƒã€‚
### é€šè¿‡ tf.random.normal(shape, mean=0.0, stddev=1.0)å¯ä»¥åˆ›å»º
### å½¢çŠ¶ä¸º shapeï¼Œå‡å€¼ä¸ºmean,æ ‡å‡†å·®ä¸º stddev çš„æ­£æ€åˆ†å¸ƒğ’©(ğ‘›ğ‘“ğ‘ğ‘œ,ğ‘¡ğ‘¢ğ‘’ğ‘’ğ‘“ğ‘¤ 2 )ã€‚
### ä¸‹é¢åˆ›å»ºä¸€ä¸ªå‡å€¼ä¸º0ï¼Œæ ‡å‡†å·®ä¸º1çš„æ­£æ€åˆ†å¸ƒ
d=tf.random.normal([2,2])
print("d:",d)
### åˆ›å»ºå‡å€¼ä¸º 1ï¼Œæ ‡å‡†å·®ä¸º 2 çš„æ­£å¤ªåˆ†å¸ƒï¼š
d1=tf.random.normal([2,2], mean=1,stddev=2)
print("d1",d1)
### é€šè¿‡ tf.random.uniform(shape, minval=0, maxval=None, dtype=tf.float32)å¯ä»¥åˆ›å»ºé‡‡æ ·è‡ª[ğ‘›ğ‘—ğ‘œğ‘¤ğ‘ğ‘š,ğ‘›ğ‘ğ‘¦ğ‘¤ğ‘ğ‘š]åŒºé—´çš„å‡åŒ€åˆ†å¸ƒçš„å¼ é‡ã€‚
### ä¾‹å¦‚åˆ›å»ºé‡‡æ ·è‡ªåŒºé—´[0,1]ï¼Œshape ä¸º[2,2]çš„çŸ©é˜µï¼š
d2=tf.random.uniform([2,2])
print("d2",d2)
### åˆ›å»ºé‡‡æ ·è‡ªåŒºé—´[0,10]ï¼Œshape ä¸º[2,2]çš„çŸ©é˜µï¼š
d3=tf.random.uniform([2,2],maxval=10)
print("d3",d3)
### å¦‚æœéœ€è¦å‡åŒ€é‡‡æ ·æ•´å½¢ç±»å‹çš„æ•°æ®ï¼Œå¿…é¡»æŒ‡å®šé‡‡æ ·åŒºé—´çš„æœ€å¤§å€¼ maxval å‚æ•°ï¼ŒåŒæ—¶åˆ¶å®šæ•°æ®ç±»å‹ä¸º tf.int*å‹ï¼š
d4=tf.random.uniform([2,2],maxval=100,dtype=tf.int32)
print("d4",d4)
## 4.4.5 åˆ›å»ºåºåˆ—
### åœ¨å¾ªç¯è®¡ç®—æˆ–è€…å¯¹å¼ é‡è¿›è¡Œç´¢å¼•æ—¶ï¼Œç»å¸¸éœ€è¦åˆ›å»ºä¸€æ®µè¿ç»­çš„æ•´å½¢åºåˆ—
### å¯ä»¥é€šè¿‡tf.range()å‡½æ•°å®ç°ã€‚
### tf.range(limit, delta=1)å¯ä»¥åˆ›å»º[0,ğ‘šğ‘—ğ‘›ğ‘—ğ‘¢)ä¹‹é—´ï¼Œæ­¥é•¿ä¸º delta çš„æ•´å½¢åºåˆ—ï¼Œä¸åŒ…å« limit æœ¬èº«ã€‚ä¾‹å¦‚ï¼Œåˆ›å»º 0~9ï¼Œæ­¥é•¿ä¸º 1 çš„æ•´å½¢åºåˆ—
e=tf.range(10,delta=1)
print("e",e)
### åˆ›å»º 0~9ï¼Œæ­¥é•¿ä¸º 2 çš„æ•´å½¢åºåˆ—ï¼š
e1=tf.range(10,delta=2)
print("e1",e1)
### tf.range(start, limit, delta=1)å¯ä»¥åˆ›å»º[ğ‘¡ğ‘¢ğ‘ğ‘ ğ‘¢,ğ‘šğ‘—ğ‘›ğ‘—ğ‘¢)ï¼Œæ­¥é•¿ä¸º delta çš„åºåˆ—ï¼Œä¸åŒ…å« limitæœ¬èº«ï¼š
e2=tf.range(1,limit=10,delta=1)
print("e2",e2)
```
